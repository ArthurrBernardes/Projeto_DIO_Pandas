```python
#Juntando todos os arquivos, tem que concatenar em um array [], ou entao não
#está certo
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
#verificando o tipo de cada coluna
df.dtypes
```

```python
#alterando o tipo da coluna LojaID para string
df["LojaID"] = df["LojaID"].astype("object")
```

```python
df.dtypes
```

```python
df.head()
```

```python
#consultando linhas de valores faltantes
df.isnull()
```

```python
#consultando linhas de valores faltantes
df.isnull().sum()
```

```python
#Se houvessem valores nulos (no video tinham 7 em vendas) vc pode substituir
#os valores nulos por médias ou por zeros
```

```python
#Substituindo pela média
df["Vendas"].fillna(df["Vendas"].mean(), inplace = True)
```

```python
df.isnull().sum()
```

```python
#Substituindo por zero
df["Vendas"].fillna(0, inplace = True)
```

```python
df.isnull().sum()
```

```python
#Apagando linhas com valores nulos
df.dropna(inplace = True)
```

```python
#Apagando linhas com valores nulos com base apenas em 1 coluna
df.dropna(subset = ["Vendas"],inplace = True)
```

```python
#Removendo linhas que estejam com alores faltantes em todas as colunas
df.dropna(how="all", inplace = True)
```

Criando Novas Colunas

```python
#Criando a coluna de receita (receita sendo vendas*quantidade)
#mul é usado para multiplicação
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
#coluna receita/vendas
df["Receita/Vendas"] = df["Receita"] / df["Vendas"]
```

```python
df.head()
```

```python
#Retornando a maior Receita
df["Receita"].max()
```

```python
#Retornando a menor Receita
df["Receita"].min()
```

```python
#Retornando as 3 maiores receitas e mostrando as lojas
df.nlargest(3, "Receita")
```

```python
#Retornando as 3 menores receitas e mostrando as lojas
df.nsmallest(3, "Receita")
```

```python
#agrupar por cidade
df.groupby("Cidade")["Receita"].sum()
```

```python
#ordenando o conjunto de dados
df.sort_values("Receita", ascending=False).head(10)
```

Trabalhando com Datas

```python
#Tranformar data para int
df["Data"] = df["Data"].astype("int64")
```

```python
#Verificando os tipos da Data (vai ser int64 inves de datetime64[ns])
df.dtypes
```

```python
#Agora tem que transformar a coluna "Data" em Data/tempo novamente
#ela so fez isso pra mostrar como faz pra transformar data em data/tempo quando
#não está na varivel correta 
df["Data"] = pd.to_datetime(df["Data"])
```

```python
df.dtypes
```

```python
#agrupamento por ano / dt é de datetime
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
#Nova coluna com Ano
df["Ano Venda"] = df["Data"].dt.year 
```

```python
df.sample(5)
```

```python
#Extraindo mês e dia
df["mes_venda"], df["dia_venda"] = (df["Data"].dt.month, df["Data"].dt.day )
```

```python
df.sample(10)
```

```python
#Data mais antiga
df["Data"].min()
```

```python
#Calculando a diferença de dias, fazendo a data de cada linha menos a 
#data minina para calcular o numero de dias (diferença de dias)
df["Diferença de dias"] = df["Data"] - df["Data"].min()
```

```python
df.sample(10)
```

```python
#Coluna de trimestre (coloca quarter pq sao 4 trimestres no ano)
df["trimestre_venda"] = df["Data"].dt.quarter
```

```python
#Obs: notar em qual trimestre está e ver como realmente os meses na coluna data
#batem com o trimestre
df.sample(10)
```

```python
#Filtrando as vendas de 2019 do mês de março
vendas_marco_19 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
```
